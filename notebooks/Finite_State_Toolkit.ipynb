###### Low-level specification

```python
myfst = FST()
```

```python
s0 = myfst.initialstate
s1 = State()
s0.add_transition(s1, ("c", "a"), 1.0)
s0.add_transition(s1, ("c", "t"), 2.0)
s1.add_transition(s0, ("c", "c"), 0.0)
s1.finalweight = 2.0
```

```python
myfst.states = {s0, s1}
myfst.finalstates = {s1}
myfst.alphabet = {"c", "a", "t"}
myfst.view()
```

```python
list(myfst.generate("ccc", weights = True))
```
